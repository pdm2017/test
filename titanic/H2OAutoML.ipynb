```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import h2o
from h2o.automl import H2OAutoML
```

```python
# JRE 설치가 필요하다
h2o.init(max_mem_size='16G')
```

```python
%%time
train = h2o.import_file("../data/train2.csv")
test = h2o.import_file("../data/test2.csv")
```

```python
y = 'Survived'
x = [x for x in train.columns if x not in y]
print(len(x), x)

train[y] = train[y].asfactor()
```

```python
#aml = H2OAutoML(max_models=30, seed=45, max_runtime_secs=18000)
aml = H2OAutoML(max_models=20, seed=45, max_runtime_secs=1800)
aml.train(x=x, y=y, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Print all rows instead of default (10 rows)
```

```python
aml.leader
```

```python
h2o.remove(train)
test = h2o.import_file("../data/test2.csv")
```

```python
preds = aml.predict(test)
preds['p1'].as_data_frame().values.flatten().shape
```

```python
print(preds)
preds.to_csv('./output/predic.csv')
```

```python
col0 = test["PassengerId"].as_data_frame().values.flatten()
col1 = preds['predict'].as_data_frame().values.flatten()

my = list(zip(col0, col1))
submit = pd.DataFrame(my, columns = ["PassengerId", "Survived"])

print(submit['Survived'].value_counts())
submit.to_csv("submissionH2O.csv", index = False) 
```
